```python
fdir_data = 'C:/Google Drive/playground/23_workspace/bug/data/accident_report/'

data = []
for fname in os.listdir(fdir_data):
    fpath = os.path.join(fdir_data, fname)
    with open(fpath, 'r', encoding='utf-8') as f:
        data.append(f.read())
```

re.findall(PATTERN, STRING) --> STRING에 등장하는 PATTERN을 모두 찾는다.

```python
re.findall(' ', data[0])
re.findall('\n', data[0])
re.findall('\s', data[0])
re.findall('\d', data[0])
re.findall('\d+', data[0])
re.findall('[ㄱ-힣]', data[0])
re.findall('[ㄱ-힣]+', data[0])
re.findall('[a-zA-Zㄱ-힣]+', data[0])
re.findall('[0-9a-zA-Zㄱ-힣]+', data[0])
re.findall('[^0-9a-zA-Zㄱ-힣]+', data[0])
```

re.search(PATTERN, STRING) --> STRING에서 가장 첫 번째로 등장하는 PATTERN의 위치(시작점, 끝점)

```python
re.search('감전', data[0])
re.search('[0-9]*[a-z]+', data[0])
```

re.sub(PATTERN, REPLACE, STRING) --> STRING에 등장하는 모든 PATTERN을 REPLACE로 변경

```python
from copy import deepcopy

tmp = deepcopy(data[0])

# tmp_sub = re.sub('감전', '피카츄', tmp)
# tmp_sub = re.sub('[^\s0-9a-zA-Zㄱ-힣]', ' ', tmp)
# tmp_sub = re.sub('[^\s0-9a-zA-Zㄱ-힣:]', ' ', tmp)
tmp_sub = re.sub('[^\s0-9a-zA-Zㄱ-힣:\.]', ' ', tmp)

print(tmp_sub)
```

re.split(PATTERN, STRING) --> STRING을 PATTERN으로 구분하여 list를 생성

※ PATTERN은 사라짐

```python
re.split('\n', data[0])
re.split('재해', data[0])
```

### Mission

1. 줄바꿈 기호('\n')를 사용하여 개별적인 문장으로 분리
2. 한글/영어/숫자, 괄호, 온점(소수점), 콜론(시간)을 제외한 불필요한 문자와 공백 제거
3. 3개 이하로 구성된 문장은 제거

```python
text = deepcopy(data[0])
text
```

```python
text_split = re.split('\n', text)
text_split
```

```python
text_remove = []
for sent in text_split:
    print('-----'*10)
    print(sent)
    
    sent_remove_special_char = re.sub('[^ 0-9a-zA-Zㄱ-힣.:()]', '', sent)
    print(sent_remove_special_char)
    
    sent_remove_space = re.sub('\s+', ' ', sent_remove_special_char)
    print(sent_remove_space)
    
    sent_remove_space_side = sent_remove_space.strip()
    print(sent_remove_space_side)
    
    text_remove.append(sent_remove_space_side)
```

```python
text_more_than_three = []
for sent in text_remove:
    print('-----'*10)
    words = re.split('\s', sent)
    print(words)
    
    if len(words) > 3:
        text_more_than_three.append(sent)
        print('--> OK')
    else:
        print('--> NO')
```
