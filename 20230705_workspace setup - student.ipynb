# Text File

파일에 접근할 때는 파일이 들어있는 폴더(fdir)와 파일의 이름(fname)을 구분하는 것이 좋음

파일의 경로(fpath)는 폴더+이름

```python
fdir_text = 'C:/Google Drive/playground/23_workspace/bug/data/accident_report'
fname_text = 'accident_0005.txt'

fpath_text = fdir_text + fname_text
print(fpath_text)
```

os: 컴퓨터와 소통하기 위한 기능들이 모여있는 기본 패키지

```python
import os

fpath_text_2 = os.path.join(fdir_text, fname_text)
print(fpath_text_2)
```

with open 구문을 사용하여 컴퓨터 폴더에 존재하는 파일을 읽어올 수 있음

- 'r': 읽겠다(read)
- encoding='utf-8': UTF-8 방식을 사용하여 읽겠다

```python
with open(fpath_text_2, 'r', encoding='utf-8') as f:
    text_data = f.read()
    
print(text_data)
```

os.listdir와 for 구문을 사용하면 여러 개의 파일을 한번에 읽어올 수 있음

```python
for fname in os.listdir(fdir_text):
    fpath = os.path.join(fdir_text, fname)
    with open(fpath, 'r', encoding='utf-8') as f:
        data = f.read()
        
    print(data)
```

미리 빈 리스트(list)를 만들어두면 여러 개의 데이터를 관리하기 수월함
- LIST_NAME.append(DATA): DATA를 LIST_NAME의 가장 끝에 이어붙인다

```python
data_list = []
for fname in os.listdir(fdir_text):
    fpath = os.path.join(fdir_text, fname)
    with open(fpath, 'r', encoding='utf-8') as f:
        data = f.read()
        
    data_list.append(data)
```

```python
data_list[7]
```

데이터의 이름을 사용하여 바로바로 접근해야 하는 경우, 딕셔너리(dict)를 사용

```python
data_dict = {}
for fname in os.listdir(fdir_text):
    fpath = os.path.join(fdir_text, fname)
    with open(fpath, 'r', encoding='utf-8') as f:
        data = f.read()
        
    data_dict[fname] = data
```

```python
data_dict['accident_0007.txt']
```

```python
for key, value in data_dict.items():
    print(key)
    print(value)
    print('--'*10)
```
